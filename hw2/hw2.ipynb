```python
# TensorFlow and tf.keras
import tensorflow as tf

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras import Input, layers

print(tf.__version__)
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
train_labels = train_labels.squeeze()

input_shape = train_images.shape[1:]
train_images = train_images / 255.0
print("Training Images range from {:2.5f} to {:2.5f}".format(np.min(train_images), np.max(train_images)))
print("Test     Images range from {:2.5f} to {:2.5f}".format(np.min(test_images), np.max(test_images)))
```

```python
model = tf.keras.Sequential([
    Input(shape=input_shape),
    layers.Conv2D(32, kernel_size=(3,3), activation="relu", padding='same', strides=(2,2)),
    layers.Conv2D(64, kernel_size=(3,3), activation="relu", padding='same', strides=(2,2)),
    layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2)),    
    layers.Flatten(),
    layers.Dense(1024, activation='relu'),
    layers.Dense(10)
])
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model.summary()
```

```python
train_hist = model.fit(train_images, train_labels, epochs=50)
#model.save('model')
```

```python
from tensorflow.keras.preprocessing import image
from keras.preprocessing.image import img_to_array

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc, '\n')

test_image = image.load_img("airplane.jpg", target_size=(32,32))
test_image = img_to_array(test_image)
test_image = np.array([test_image]) / 255.0
#print(test_image)

pred = model.predict(test_image, batch_size=1)
print(pred)
pred_class = model.predict_classes(test_image, batch_size=1)
print(pred_class)
```

```python
new_model = tf.keras.Sequential([
    Input(shape=input_shape),
    layers.DepthwiseConv2D(kernel_size=(1,1)),
    layers.Conv2D(32, kernel_size=(1,1), activation="relu"),
    layers.DepthwiseConv2D(kernel_size=(1,1)),
    layers.Conv2D(64, kernel_size=(1,1), activation="relu"),
    layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2)),    
    layers.Flatten(),
    layers.Dense(1024, activation='relu'),
    layers.Dense(10)
])

new_model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
new_model.summary()
```

```python
new_train_hist = new_model.fit(train_images, train_labels, epochs=50)
#new_model.save('new_model')
```

```python
test_loss, test_acc = new_model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc, '\n')

test_image = image.load_img("airplane.jpg", target_size=(32,32))
test_image = img_to_array(test_image)
test_image = np.array([test_image]) / 255.0
#print(test_image)

pred = new_model.predict(test_image, batch_size=1)
print(pred)
pred_class = new_model.predict_classes(test_image, batch_size=1)
print(pred_class)
```
